```python
def predict(weights, corrupted_images, patterns_count, height, width):
    result = np.zeros((patterns_count, 1, height*width))
    for i in range(patterns_count):
        
        result[i] = corrupted_images[i]
        prev_value = np.copy(result[i])

        energy = get_energy(result[i].T, weights)
        while True:
            if (np.array_equal(result[i], activate(result[i] @ weights, prev_value))):
                break;
            
            result[i] = activate(result[i] @ weights, prev_value)
            
            prev_value = np.copy(result[i])
            new_energy = get_energy(result[i].T, weights)
            if (np.array_equal(energy, new_energy)):
                break
            energy = new_energy
            
    return result

def activate(data, prev_value):
    copied_data = np.copy(data)
    copied_data = np.sign(copied_data)
    
    copied_data[copied_data == 0] = prev_value[copied_data == 0]
    return copied_data

def train(patterns_count, weights, data):
    copied_weights = np.copy(weights)
    for i in range(patterns_count):
        t1 = copied_weights @ data[i].T - data[i].T
        t2 = t1.T
        coeff = 1 / (data[i] @ data[i].T - data[i] @ copied_weights @ data[i].T)
        copied_weights += coeff * t1 * t2
    diagW = np.diag(np.diag(copied_weights))
    copied_weights = copied_weights - diagW
    
    return copied_weights

def show_image(initial_image, corrupted_image, restored_image):
         fig, axs = plt.subplots(3, figsize=(5, 15))
         axs[0].imshow(initial_image, cmap='Greys_r') 
         axs[1].imshow(corrupted_image, cmap='Greys_r')
         axs[2].imshow(restored_image, cmap='Greys_r')
         plt.show()

def get_energy(data, weights):
    return -0.5 * data.T @ weights @ data
    
height, width, patterns_count = images.shape[1], images.shape[2], images.shape[0]

images = np.array(images).reshape((patterns_count, 1, height * width))
corrupted_images = np.array(corrupted_images).reshape((patterns_count, 1, height * width))

weights = np.zeros((height*width, height*width))

weights = train(patterns_count, weights, images)
result = predict(weights, corrupted_images, patterns_count, height, width)

result = result.reshape((patterns_count, height, width))
images = images.reshape((patterns_count, height, width))
corrupted_images = corrupted_images.reshape((patterns_count, height, width))
for i in range(patterns_count):
    show_image(images[i], corrupted_images[i], result[i])
    
```
